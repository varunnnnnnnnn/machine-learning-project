```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("marketing_campaign.csv")
```

```python
df_filtered = df.dropna(subset=['Income'])
df_filtered.head()
```

```python
from certifi import where


df_numeric=df_filtered.select_dtypes(int,float)
df_numeric
```

```python
from sklearn.preprocessing import normalize
data_scaled = normalize(df_numeric)
data_scaled = pd.DataFrame(data_scaled, columns=df_numeric.columns)
data_scaled.head()
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(data_scaled, method='ward'))
```

```python
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(data_scaled, method='ward'))
plt.axhline(y=8, color='r', linestyle='--')
```

Number of clusters=3

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, linkage='ward')  
cluster.fit_predict(data_scaled)
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(data_scaled['Year_Birth'], data_scaled['NumWebVisitsMonth'], c=cluster.labels_)
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(data_scaled['Recency'], data_scaled['NumWebVisitsMonth'], c=cluster.labels_)
```
