```python
upper_limit = df_filtered['Income'].mean() + 3*df_filtered['Income'].std()
lower_limit = df_filtered['Income'].mean() - 3*df_filtered['Income'].std()
df_filtered['Income'] = np.where(df_filtered['Income']>upper_limit,upper_limit,np.where(df_filtered['Income']<lower_limit,lower_limit,df_filtered['Income']))
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.scatter(x = df_filtered['age'], y = df_filtered['Income'])
plt.ylabel('Age', fontsize=13)
plt.xlabel('Income', fontsize=13)
plt.show()
```

```python
import seaborn as sns
from scipy.stats import norm, skew
from scipy import stats
sns.distplot(df_filtered['Income'] , fit=norm)

# Get the fitted parameters used by the function
(mu, sigma) = norm.fit(df_filtered['Income'])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))
#We use the numpy fuction log1p which  applies log(1+x) to all elements of the column
df_filtered['Income'] = np.log1p(df_filtered['Income'])

#Check the new distribution 
sns.distplot(df_filtered['Income'] , fit=norm)

# Get the fitted parameters used by the function
(mu, sigma) = norm.fit(df_filtered['Income'])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

#Now plot the distribution
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequency')
plt.title('Income')

#Get also the QQ-plot
fig = plt.figure()
res = stats.probplot(df_filtered['Income'], plot=plt)
plt.show()
```

The data is leftly skewed indicating a left-skewed target variable has most data points concentrated on the right side, with a tail extending towards lower values, that is most of the income values fall on the upper side,indicating higher salary.

```python
#create a new column for age
df_filtered['age'] = df_filtered['Year_Birth'].apply(lambda x: 2024-x)
model = LinearRegression()
df_filtered = df_filtered.dropna(subset=['Income'])  # Drops rows with missing income values
X = df_filtered['age'].values.reshape(-1, 1)
y = df_filtered['Income']
model.fit(X, y)
```

```python
r_sq = model.score(X, y)
print(f"coefficient of determination: {r_sq}")
```

```python
print(f"intercept: {model.intercept_}")
print(f"slope: {model.coef_}")
```
